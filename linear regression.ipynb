```python
import numpy as np  #import numpy as np
import pandas as pd
```

```python
train=np.genfromtxt('train_data.csv',delimiter=",")  #reading train and test data from csv file to np arrays

train=np.delete(train,0,axis=0)                      #axis=0 denotes row. hence deletes 0th row(row of Nan(not integers)
```

```python
m=len(train)  #no of training examples
label=train[:,-1]  #labels for each training data
features=train[:,0:4]  # features x1 to x4 in the training set
features=np.insert(features,0,np.ones(m),axis=1)  #adding x0=1 to our feature matrix
lr=.01  #learning rate alpha=.01
theta=np.zeros(5)        #initializing all parameters to zero
count=1
```

```python
def cost(hyp,label):                             #cost calculation function
    summation=np.sum((hyp-label)**2)             #sum of square of difference (hx-y)
    cost=summation/2*m                              #calculation of cost by finding the mean of error                    
    return cost
```

```python
hyp=np.dot(features,theta)         #initial hypothesis by keeping all parameters equal to 0  m*1(no need to transpose 1Darrays)
initialcost=cost(hyp,label)                     #initial cost
difference=hyp-label
```

```python
while(True):
    theta=theta-(lr/m)*np.dot(difference.transpose(),features)
    hyp=np.dot(features,theta)
    difference=hyp-label
    newcost=cost(hyp,label)
    print "serial no:",count,"      value: ",newcost
    count+=1
    if abs(newcost-initialcost)<=.0000000000000001:
        break
    else:
        initialcost=newcost    
```

```python
print"final parameters",theta
```

```python
print"difference between actual values and prediction accoding to hypothesis"
print"predicted\t"
predicted= np.dot(features,theta)
print pd.DataFrame({'predicted':[predicted],'actual':[label]})
```

```python
print"error in each prediction"
print pd.DataFrame(data=predicted-label,columns=["error"])
```
